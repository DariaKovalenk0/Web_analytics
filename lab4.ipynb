```python
import pandas as pd
from GoogleNews import GoogleNews

def news(target, start_date, end_date, max_results=2000):
    data = GoogleNews()
    data.set_time_range(start_date, end_date)

    all_results = []
    page = 1

    while len(all_results) < max_results:
        data.search(target)
        results = data.results()

        if not results:
            break

        all_results.extend(results)
        page += 1
        data.getpage(page)

    df = pd.DataFrame(all_results[:max_results])
    df.to_csv(f'{target}.csv', index=False)

    return df

data = news('Ukraine', '01/01/2023', '12/31/2023')
data.head()
```
